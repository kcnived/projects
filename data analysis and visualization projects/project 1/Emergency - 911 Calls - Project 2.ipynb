```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head(3)
```

```python
df['zip'].value_counts().head(5) #top 5 zipcodes for 911 calls
```

```python
df['twp'].value_counts().head(5) # top 5 townships (twp) for 911 calls
```

```python
df['title'].nunique() #unique title codes
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0]) #a new column called "Reason" that contains this string value
```

```python
df['Reason'].value_counts() #most common Reason for a 911 call 
```

```python
sns.countplot(x='Reason',data=df,palette='viridis') # countplot of 911 calls by Reason
```

```python
type(df['timeStamp'].iloc[0]) #data type of the objects in the timeStamp column
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp']) #convert the column from strings to DateTime objects
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek) #3 new columns called Hour, Month, and Day of Week
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'} # map the actual string names to the day of the week:
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
# a countplot of the Day of Week column with the hue based off of the Reason column
sns.countplot(x='Day of Week',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#group the DataFrame by the month column and use the count() method for aggregation.
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
# count of calls per month
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
df['Date']=df['timeStamp'].apply(lambda t: t.date()) #a new column called 'Date' that contains the date from the timeStamp column.
```

```python
df.groupby('Date').count()['twp'].plot()
plt.tight_layout() # groupby this Date column with the count() aggregate and created a plot of counts of 911 calls.
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
#restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week.
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
sns.clustermap(dayHour,cmap='viridis')
```

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
sns.clustermap(dayMonth,cmap='viridis')
```
